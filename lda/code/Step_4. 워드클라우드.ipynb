```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
wordcloud_vocab = pd.read_pickle(r"C:\Users\user\Desktop\새 폴더\data\wordcliud_dict.pkl")
```

```python
wc = WordCloud(random_state = 123, font_path = 'NanumGothic', width = 400,
               height = 400, background_color = 'white').generate_from_frequencies(wordcloud_vocab)
```

```python
plt.figure()
plt.imshow(wc)
plt.axis("off")
```
